```python
# Importing necessary libraries

import pandas as pd            
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sb 
```

```python
# Reading csv files into dataframes df and df1

df= pd.read_csv('heatmapcorrelationhealth.csv')
df1= pd.read_csv('heatmapcorrelation.csv')
```

```python
# Calculating correlation matrix 1

corrmat = df.corr(numeric_only=True)
corrmat
```

```python
# Create a figure and axis for the heatmap
fig, ax = plt.subplots(figsize=(10,6))

# Generate a heatmap using seaborn, specifying the correlation matrix (corrmat)
# Set the color map to "coolwarm" and disable annotations (annot=False)
sb.heatmap(corrmat,ax=ax,cmap="coolwarm",annot=False)

# Add a title to the heatmap
plt.title("Correlation Heatplot of Education and Self-Rated Health Evaluation")

# Save the heatmap as an image named "heatmaphealth" with a DPI of 300
# Ensure tight layout for proper image dimensions
plt.savefig("heatmaphealth",dpi=300,
           bbox_inches="tight",)
```

```python
# Calculating correlation matrix 2

corrmat1 = df1.corr(numeric_only=True)
corrmat1
```

```python

# Create new figure and axis for the heatmap
fig1, ax1 = plt.subplots(figsize=(10,6))

# Generate a heatmap using seaborn, specifying second correlation matrix (corrmat1)
# Set the color map to "coolwarm" and disable annotations (annot=False)
sb.heatmap(corrmat1,ax=ax1,cmap="coolwarm",annot=False)

# Add a title to the heatmap
plt.title("Correlation Heatplot of Education and Household Income/IMD")

# Save the heatmap as an image named "heatmapedu" with a DPI of 300
# Ensure tight layout for proper image dimensions
plt.savefig("heatmapedu",dpi=300,
           bbox_inches="tight",)
```
